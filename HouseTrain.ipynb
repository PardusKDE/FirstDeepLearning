```
from keras.layers import Dense
from keras.models import Sequential
```

```
import pandas as pd
```

```
csv = pd.read_csv("sample_data/california_housing_train.csv")
```

```
csv.head()
```

```
X = csv["longitude"]
y = csv["latitude"]
```

```
X.head()
```

```
y
```

```
model = Sequential()
```

```
model.add(Dense(128))
model.add(Dense(50))
```

```
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy")
```

```
model.fit(X, y, epochs=30)
```

```
acc = model.evaluate(X, y)
```

```
acc
```

```
pd.DataFrame(model.predict(y))
```
